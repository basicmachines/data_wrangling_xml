## XML example

+ for details about tree traversal and iterators, see https://docs.python.org/2.7/library/xml.etree.elementtree.html

```python
document_tree = ET.parse( './data/mondial_database_less.xml' )
```

```python
# print names of all countries
for child in document_tree.getroot():
    print child.find('name').text
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):
    print '* ' + element.find('name').text + ':',
    capitals_string = ''
    for subelement in element.getiterator('city'):
        capitals_string += subelement.find('name').text + ', '
    print capitals_string[:-2]
```

Using data in 'data/mondial_database.xml', the examples above, and refering to https://docs.python.org/2.7/library/xml.etree.elementtree.html, find

```python
# Find 10 countries with highest infant mortality

data = []

print "Infant Mortality"
for country in document_tree.iterfind('country'):
    name = country.find('name')
    value = country.find('infant_mortality')
    if value is not None:
        item = (name.text, float(value.text))
        print '%22s: %s' % item
        data.append(item)
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['country_name', 'infant_mortality'])
df.head()
```

```python
for city in root[0].findall('city'):
    print city.find('name').text
    population_data = city.findall('population')
    n = len(population_data)
    print n, "population figures found"

    if n>0:
    
        most_recent_year = 0
        pop_choice = None

        for pop in population_data:
            measured = pop.get('measured')
            year = pop.get('year')
            if year > most_recent_year:
                most_recent_year = year
                pop_choice = pop

        print "Most recent: %s (%s %s)\n" % (pop_choice.text, pop_choice.get('year'), pop_choice.get('measured'))

```

```python
# Find the 10 cities with the largest population

data = []

print "Population of cities"
for country in document_tree.iterfind('country'):
    for city in country.iterfind('city'):
        name = city.find('name')

        population_data = city.findall('population')
        
        if len(population_data) > 0:

            most_recent_year = 0
            value = None

            for pop in population_data:
                year = pop.get('year')
                if year > most_recent_year:
                    most_recent_year = year
                    value = pop

            item = (name.text, int(value.text))
            print '%22s: %s' % item
            data.append(item)
```

```python
df = pd.DataFrame(data, columns=['city_name', 'population'])
df.head()
```

```python
df.sort_values('population', ascending=False).head(10)
```

```python
for egroup in root[0].iter('ethnicgroup'):
    for element in egroup.iter():
        print element.tag, element.attrib, element.text
```

```python
ethnic_group_counts = dict()
for country in root.findall('country'):
    print country.find('name').text
    
    population_data = country.findall('population')
    n = len(population_data)

    if n>0:
    
        most_recent_year = 0
        pop_choice = None

        for pop in population_data:
            measured = pop.get('measured')
            year = pop.get('year')
            if year > most_recent_year:
                most_recent_year = year
                pop_choice = pop

        print " Most recent population: %s (%s %s)" % (pop_choice.text, pop_choice.get('year'), pop_choice.get('measured'))

        population = int(pop_choice.text)
        
        count = 0
        for egroup in country.iter('ethnicgroup'):
            for element in egroup.iter():
                print "   - %s: %s %s" % (element.tag, element.text, element.get('percentage'))
                ethnic_group_counts[element.text] = ethnic_group_counts.get(element.text, 0) + population
                count += 1
        
        if count == 0:
            ethnic_group_counts['Unknown'] = ethnic_group_counts.get('Unknown', 0) + population
```

```python
# Note, no data for some countries. E.g. France, Italy, Denmark, Venezuela, ...
ethnic_group_counts['Unknown']
```

```python
df = pd.DataFrame.from_dict(ethnic_group_counts, orient='index')
df.columns=['population']
```

```python
def max_attribute_value(root, element_string, attribute_string, other_attribute_string=None):
    
    largest = (None, None)
    
    for element in root.findall(element_string):
        
        other_attribute = None
        
        element_name = element.find('name').text
        if other_attribute_string is not None:
            other_attribute = element.find(other_attribute_string)
        
        if element.find(attribute_string) is not None:
            if element.find(attribute_string).text is not None:
                try:
                    attribute_value = int(element.find(attribute_string).text)
                except:
                    try:
                        attribute_value = float(element.find(attribute_string).text)
                    except:
                        print "Error: attribute could not be converted to a number", element.find(attribute_string).text
                #print " -", element_name, attribute_value

            if attribute_value > largest[1]:
                largest = (element_name, attribute_value)

                if other_attribute is not None:
                    largest = (largest[0], largest[1], other_attribute)
    
    return largest


longest_river = max_attribute_value(root, "river", "length", other_attribute_string="located")
print "Longest river:", longest_river, longest_river[2].attrib

largest_lake = max_attribute_value(root, "lake", "area") #, other_attribute_string="located")
print "Largest lake:", largest_lake

airport_at_highest_elevation = max_attribute_value(root, "airport", "elevation")
print "Airport at highest elevation:", airport_at_highest_elevation


```
